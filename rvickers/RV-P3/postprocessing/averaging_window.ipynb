```python
gap_sizes = ['06','10','20','30']
pressure_diffs = ['100','200','500','1000']
averaging_windows = [(4e6,6.5e6),(3e6,6.5e6),(3e6,6.5e6),(3e6,6.5e6)]
colors = ['blue','green','red','purple']

for gap_size,averaging_window in zip(gap_sizes,averaging_windows):
    pressureDFs = []
    flowrates = []
    deltaps = []
    p_fig,p_ax = plt.subplots()
    d_fig,d_ax = plt.subplots()
    v_fig,v_ax = plt.subplots()
    pgap_fig,pgap_ax = plt.subplots()
    vgap_fig,vgap_ax = plt.subplots()
    pcount_fig,pcount_ax = plt.subplots()
    flow_fig,flow_ax = plt.subplots()
    for pressure_diff, color in zip(pressure_diffs, colors):
        pressureDF = pd.read_csv('./csv/pagap'+gap_size+'_nve'+pressure_diff+'_pressure_avg.csv')
        pressureDF = pressureDF.iloc[20:-10]
        pfeed = pressureDF['mean'].iloc[0:70].mean()
        pperm = pressureDF['mean'].iloc[-40:-1].mean()
        pressureDF.rename(columns={'Distance in z ()' : 'Distance in z ()', 'mean' : '$\Delta$P = '+pressure_diff, 'stdev' : pressure_diff+'stdev'},inplace=True)
        pressureDF.plot(x='Distance in z ()', y='$\Delta$P = '+pressure_diff, ylabel='Pressure of molecules within bin (ATM)', xlim=(-50, 375), ylim = (-2000,1500),ax=p_ax, color=color)
        p_ax.errorbar(pressureDF['Distance in z ()'], pressureDF['$\Delta$P = '+pressure_diff], yerr = pressureDF[pressure_diff+'stdev'], alpha=0.05, color = color)
    
        densityDF = pd.read_csv('./csv/pagap'+gap_size+'_nve'+pressure_diff+'_density_avg.csv')
        densityDF = densityDF.iloc[20:-10]
        densityDF.rename(columns={'Distance in z ()' : 'Distance in z ()', 'mean' : '$\Delta$P = '+pressure_diff, 'stdev' : pressure_diff+'stdev'},inplace=True)
        densityDF.plot(x='Distance in z ()', y='$\Delta$P = '+pressure_diff, ylabel='Density of molecules within bin (g/cm$^3$)',xlim=(-50, 375), ylim = (0.9,1.1),ax=d_ax, color=color)
        d_ax.errorbar(densityDF['Distance in z ()'], densityDF['$\Delta$P = '+pressure_diff], yerr = densityDF[pressure_diff+'stdev'], alpha=0.05, color = color)
        
        velocityDF = pd.read_csv('./csv/pagap'+gap_size+'_nve'+pressure_diff+'_vz_avg.csv')
        velocityDF = velocityDF.iloc[20:-10]
        velocityDF.rename(columns={'Distance in z ()' : 'Distance in z ()', 'mean' : '$\Delta$P = '+pressure_diff, 'stdev' : pressure_diff+'stdev'},inplace=True)
        velocityDF.plot(x='Distance in z ()',y='$\Delta$P = '+pressure_diff, ylabel='Velocity of molecules in direction of flow (/fs)',xlim=(-50, 375),ylim = (-3e-4,3e-4),ax=v_ax, color=color)
        v_ax.errorbar(velocityDF['Distance in z ()'], velocityDF['$\Delta$P = '+pressure_diff], yerr = velocityDF[pressure_diff+'stdev'], alpha=0.05, color = color)
        
        pressureGapDF = pd.read_csv('./csv/pagap'+gap_size+'_nve'+pressure_diff+'_pressure_gap_avg.csv')
        pressureGapDF.rename(columns={'Distance in y ()' : 'Distance in y ()', 'mean' : '$\Delta$P = '+pressure_diff, 'stdev' : pressure_diff+'stdev'},inplace=True)
        pressureGapDF.plot(x='Distance in y ()',y='$\Delta$P = '+pressure_diff, ylabel='Pressure of molecules within bin (ATM)',ax=pgap_ax, color=color)
        pgap_ax.errorbar(pressureGapDF['Distance in y ()'], pressureGapDF['$\Delta$P = '+pressure_diff], yerr = pressureGapDF[pressure_diff+'stdev'], alpha=0.05, color = color)
        
        velocityGapDF = pd.read_csv('./csv/pagap'+gap_size+'_nve'+pressure_diff+'_vz_gap_avg.csv')
        velocityGapDF.rename(columns={'Distance in y ()' : 'Distance in y ()', 'mean' : '$\Delta$P = '+pressure_diff, 'stdev' : pressure_diff+'stdev'},inplace=True)
        velocityGapDF.plot(x='Distance in y ()',y='$\Delta$P = '+pressure_diff, ylabel='Velocity of molecules in direction of flow (/fs)',ylim = (-3e-4,3e-4),ax=vgap_ax, color=color)
        vgap_ax.errorbar(velocityGapDF['Distance in y ()'], velocityGapDF['$\Delta$P = '+pressure_diff], yerr = velocityGapDF[pressure_diff+'stdev'], alpha=0.05, color = color)
        
        existingCountResDF = pd.read_csv('./csv/pagap'+gap_size+'_nve'+pressure_diff+'_res_counts.csv',index_col=0)
        existingCountResDF = existingCountResDF.rolling(50,axis=1).mean()
        initCountResDF = existingCountResDF.loc[['perm'], [col for col in existingCountResDF.columns if ((int(col)>averaging_window[0])&(int(col)<averaging_window[1]))]].iloc[:,0]
        deltaCountResDF = existingCountResDF.loc[['perm'], [col for col in existingCountResDF.columns if ((int(col)>averaging_window[0])&(int(col)<averaging_window[1]))]].sub(initCountResDF,axis=0)
#         deltaCountResDF.index.rename('$\Delta$P = '+pressure_diff,inplace=True)
        linregr = LinearRegression(fit_intercept=False)
        x = deltaCountResDF.T.index.values.reshape(len(deltaCountResDF.T.index.values),1).astype(int)
        x = x-x[0]
        y = deltaCountResDF.T['perm'].values.reshape(len(deltaCountResDF.T['perm'].values),1)
        linregr.fit(x,y)
        flowrates.append((linregr.coef_*18/avo_num/2*1e15)[0][0])
        deltaps.append(pfeed-pperm)
        deltaCountResDF = deltaCountResDF.T
        deltaCountResDF['fit'] = linregr.predict(x)
        deltaCountResDF.plot(y='fit',legend=False,ax=pcount_ax,color=color,style='--')
        deltaCountResDF.plot(y='perm',label= '$\Delta$P = '+pressure_diff,xlabel='timestep',ylabel='Change in number of water molecules\n within permeate region',legend=True,ax=pcount_ax,color=color)
    p_ax.set_xlabel('Distance in z ()')
    p_ax.set_title('Molecular pressure of water flowing through '+ gap_size +'  gap\nBinned normal to flow')
    d_ax.set_xlabel('Distance in z ()')
    d_ax.set_title('Density of water flowing through '+ gap_size +'  gap\nBinned normal to flow')
    v_ax.set_xlabel('Distance in z ()')
    v_ax.set_title('Velocity of water flowing through '+ gap_size +'  gap\nBinned normal to flow')
    
    pgap_ax.set_xlabel('Distance in y ()')
    pgap_ax.set_title('Molecular pressure of water within '+ gap_size +'  gap\nBinned in direction of flow')
    
    vgap_ax.set_xlabel('Distance in y ()')
    vgap_ax.set_title('Velocity of water within '+ gap_size +'  gap\nBinned in direction of flow')
    
    pcount_ax.set_title('Change in number of water molecules in permeate during Steady State')
    
    flow_ax.plot(deltaps,flowrates,marker='o',linestyle='None',color='blue')
    x = np.array(deltaps).reshape(len(deltaps),1)
    y = np.array(flowrates).reshape(len(flowrates),1)
    linregr = LinearRegression(fit_intercept=False)
    linregr.fit(x,y)
    
    flow_ax.plot(x,linregr.predict(x),linestyle='--',color='blue')
    flow_ax.set_xlabel('Pressure difference across sample (ATM)')
    flow_ax.set_ylabel('Mass flow rate through gap (g/s)')
    flow_ax.set_title('Mass flow rate of water through '+gap_size+'  gap')
    flow_ax.text(0.33,0.67,'R$^2$ = '+str(round(linregr.score(x,y),3)), 
                 verticalalignment='center', horizontalalignment='center',
                 fontsize=12,transform=flow_ax.transAxes)
    
    p_fig.savefig('./plots/pagap'+gap_size+'_nve_pressure_avg_aggregate.jpg',bbox_inches='tight')
    d_fig.savefig('./plots/pagap'+gap_size+'_nve_density_avg_aggregate.jpg',bbox_inches='tight')
    v_fig.savefig('./plots/pagap'+gap_size+'_nve_vz_avg_aggregate.jpg',bbox_inches='tight')
    pgap_fig.savefig('./plots/pagap'+gap_size+'_nve_pressure_gap_avg_aggregate.jpg',bbox_inches='tight')
    vgap_fig.savefig('./plots/pagap'+gap_size+'_nve_vz_gap_avg_aggregate.jpg',bbox_inches='tight')
    pcount_fig.savefig('./plots/pagap'+gap_size+'_nve_perm_counts_aggregate.jpg',bbox_inches='tight')
    flow_fig.savefig('./plots/pagap'+gap_size+'_nve_flow_rates.jpg',bbox_inches='tight')
```
